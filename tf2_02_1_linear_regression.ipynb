```
import numpy as np
import tensorflow as tf
```

```
x_train = [1,2,3,4]
y_train = [0,-1,-2,-3]
```

```
tf.model = tf.keras.Sequential()
tf.model.add(tf.keras.layers.Dense(units=1, input_dim=1)) #units == output shape, input_dim==input shape
```

```
sgd = tf.keras.optimizers.SGD(lr=0.1) # SGD = standard gradient decent
tf.model.compile(loss='mse', optimizer=sgd)  #평균제곱오차MSL = mean square error
```

```
# prints summary of the model to the terminal
tf.model.summary
```

```
tf.model.fit(x_train, y_train, epochs=200)
```

```
y_predict = tf.model.predict(np.array([4,5]))  # 이 부분을 모르겠네..?
y_predict
```
